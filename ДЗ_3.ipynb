````python
# Создадим датафреймы
df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie','David','Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Score': [88, 92, 85, 90, 95]
})

df
````

````python
df2 = pd.DataFrame({
    'Dept': ['HR','HR','IT','IT','HR'],
    'Salary': [100, 150, 200, 250, 180]
})

df2
````

````python
# Отфильтруем записи в df по условию Age > 30 и Score >= 90
filtered_df = df[(df['Age'] > 30) & (df['Score'] >= 90)]
filtered_df
````

````python
# Cозданим DataFrame с покупками

purchases_df = pd.DataFrame({
    'User': ['Иван', 'Мария', 'Иван', 'Петр', 'Мария', 'Иван', 'Анна', 'Петр'],
    'Item': ['яблоко', 'банан', 'апельсин', 'яблоко', 'груша', 'банан', 'яблоко', 'апельсин'],
    'Quantity': [2, 3, 1, 4, 2, 2, 3, 1]
})

purchases_df
````

````python
# Подсчитаем количество товаров по каждому пользователю
user_purchases = purchases_df.groupby('User')['Quantity'].sum().to_frame()
user_purchases
````

````python
# Отсортируем df2 по Dept и Salary по возрастанию
sorted_df2 = df2.sort_values(['Dept', 'Salary'], ascending=[True, True])

sorted_df2
````

````python
# Создадим датафрейм с данными об возрастах студентов из разных классов
students_df = pd.DataFrame({
    'Student': ['Анна', 'Борис', 'Виктория', 'Георгий', 'Дарья', 'Евгений', 'Жанна', 'Захар'],
    'Age': [16, 17, 16, 18, 16, 16, 18, 17],
    'Class': ['10A', '10Б', '10A', '11A', '10Б', '11A', '10A', '11Б']
})

students_df
````

````python
# Вычислим средний возраст студентов в каждом классе
avg_age_by_class = students_df.groupby('Class')['Age'].mean().to_frame()

avg_age_by_class
````

- В вашем DataFrame найдите количество пропущенных значений в каждом столбце (используйте

- Удалите все строки, где присутствуют пропуски во всех столбцах одновременно.
- Создайте столбец с датами в строковом формате (например, `df['date_str'] =

- Приведите числовой столбец с плавающей точкой к целочисленному типу, а строковый столбец с
числами – к числовому типу с помощью pd.to_numeric()

````python
# Создаем DataFrame с пропущенными значениями для демонстрации
df_with_na = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', np.nan],
    'Age': [25, np.nan, 35, 40, 22, np.nan],
    'Score': [88, 92, np.nan, 90, 95, np.nan],
    'Grade': [np.nan, 'B', 'C', 'A', 'A', np.nan]
})

df_with_na
````

````python
# Нахождение количества пропущенных значений в каждом столбце
missing_values = df_with_na.isna().sum()

missing_values
````

````python
# Удаляем строки, где все значения NaN
df_cleaned = df_with_na.dropna(how='all')

df_cleaned
````

````python
# Преобразование строкового столбца с датами в datetime и извлечение года и месяца
dates_df = pd.DataFrame({
    'date_str': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05', '2023-05-12']
})

dates_df
````

````python
# Преобразуем строковый столбец в формат datetime
dates_df['date_datetime'] = pd.to_datetime(dates_df['date_str'])

dates_df
````

````python
dates_df.info()
````

````python
# Извлекаем год и месяц
dates_df['year'] = dates_df['date_datetime'].dt.year
dates_df['month'] = dates_df['date_datetime'].dt.month

dates_df
````

````python
# Преобразование типов данных
types_df = pd.DataFrame({
    'float_col': [1.5, 2.7, 3.9, 4.2, 5.8],  # Столбец с плавающей точкой
    'str_num_col': ['10', '20', '30', '40', '50']  # Столбец со строками, содержащими числа
})

types_df
````

````python
# Типы данных
types_df.dtypes
````

````python
# Преобразуем столбец с плавающей точкой к целочисленному типу
types_df['int_col'] = types_df['float_col'].astype(int)

# Преобразуем строковый столбец с числами к числовому типу с помощью pd.to_numeric()
types_df['numeric_col'] = pd.to_numeric(types_df['str_num_col'])

types_df
````
